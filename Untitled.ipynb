```python
test=pd.read_csv("test.csv")
```

```python
test=test.drop(['PoolQC','Fence','MiscFeature','FireplaceQu','Alley','Id'],axis=1)
```

```python
objects=test.select_dtypes(['object']).columns
```

```python
for c in objects:
    test[c]=test[c].fillna(test[c].mode()[0])
   
         
```

```python
floats=test.select_dtypes(['float64']).columns
```

```python
for c in floats:
    test[c]=test[c].fillna(test[c].mean())
```

```python
price=df['SalePrice']
df=df.drop('SalePrice',axis=1)
```

```python
data_set=pd.concat([df,test],axis=0)
print(data_set.shape)
```

```python
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
encoder=OneHotEncoder(sparse=False,drop='first')
```

```python
object_col=data_set.select_dtypes(include=['object']).columns
object_col
```

```python
dummy=pd.DataFrame(encoder.fit_transform(data_set[object_col]))
dummy.columns=encoder.get_feature_names(object_col)
dummy.head()
```

```python
data_set=data_set.drop(object_col,axis=1)
data_set
```

```python
data_set=data_set.reset_index(drop='True')
```

```python
data_set=pd.concat([data_set,dummy],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
scaler.fit(data_set)
```

```python
data_set=pd.DataFrame(scaler.transform(data_set))
```

```python
train=data_set[:len(df)]
test=data_set[len(df):]
print(train.shape)
print(test.shape)
```

```python
X_train, X_test, y_train, y_test=train_test_split(train,price,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.metrics import mean_squared_error
```

```python
model=keras.Sequential(
[   layers.BatchNormalization(input_shape=[X_train.shape[1]]),
    layers.Dense(265,activation='relu'),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
 
 layers.Dense(265,activation='relu'),
    layers.Dropout(0.3),
    layers.BatchNormalization(),

    layers.Dense(128,activation='relu'),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
 
    layers.Dense(64,activation='relu'),
 layers.Dropout(0.3),
    layers.BatchNormalization(),
    layers.Dense(1)
    
])
```

```python
model.compile(
loss='mean_squared_error',
optimizer='adam',
metrics=['accuracy']
)
```

```python
early_stopping=keras.callbacks.EarlyStopping(
patience=25,
min_delta=0.01,
restore_best_weights=True
)
```

```python
history=model.fit(
X_train, y_train,
validation_data=(X_test,y_test),
batch_size=64,
epochs=500,
callbacks=[early_stopping]
)
```

```python
sklearn.metrics.r2_score(model.predict(X_test),y_test)
```

```python
np.sqrt(mean_squared_error(model.predict(X_test),y_test))
```

```python
history.history.keys()
```

```python
history_df=pd.DataFrame(history.history)
history_df.loc[ : ,['loss','val_loss']].plot()
```
